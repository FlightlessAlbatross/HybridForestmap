use the vrt to create a reprojected output file. Also cut the output into blocks based on the output CRS
the coordinates xmin xmax etc are in geocoordinates, we get those from the CLC 10m chunks

```python
%%bash
start=`date +%s`


for i in {2..4..1} ;do
for j in {1..2..1} ;do

xstart=$((  900000  + (i-1) * 1625000 ))
ystart=$((  5500000 - j     * 1150000 ))
xend=$((    900000  + i     * 1625000 ))
yend=$((    5500000 - (j-1) * 1150000 ))

outfile=in_data/GLADGLCextent/reproj3035_tifs/reproj3035_$i$j.tif

echo "Start: "$outfile

gdalwarp -co COMPRESS=DEFLATE -co ZLEVEL=9 \
-overwrite                                 \
-t_srs EPSG:3035                           \
-te $xstart $ystart $xend $yend            \
-te_srs EPSG:3035                          \
-dstnodata 48                              \
in_data/GLADGLCextent/tifs/EUextent.vrt $outfile

echo "End: "$outfile

done
done

end=`date +%s`
expr $end - $start
```
