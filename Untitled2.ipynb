```python
import rasterio
import geopandas as gpd
from rasterstats import zonal_stats
from shapely.geometry import box
```

```python
coarser_poly_path = 'in_data/reference_1km/Poland/pl_1km.shp'
raster_path = 'out_data/predictions/landcover_fillup/PL72_prediction_kernel.tif'
```

```python

# Read the finer raster
with rasterio.open(raster_path) as src:
    finer_raster = src.read(1)  # read the first band
    finer_transform = src.transform
    raster_bounds = src.bounds

raster_bbox = box(*raster_bounds)
    
coars_poly = gpd.read_file(coarser_poly_path)

```

```python
# Check which polygons are contained within the raster's bounding box
contained_polygons = coars_poly[coars_poly.geometry.within(raster_bbox)]

contained_polygons = contained_polygons.copy()
```

```python
stats = zonal_stats(contained_polygons, finer_raster, affine=finer_transform, categorical=True)
```

```python
for cl in [0,1,2,3,4,255]:
    l = []
    for cell in stats:
        
        
        if cl in cell.keys():
            l.append(cell[cl] / 100)
        else:
            l.append(0)
    
    contained_polygons.loc[:,"cl" + str(cl)] = l
```
