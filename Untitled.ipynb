```python
%%bash

for i in {3..3..1} ;do
for j in {3..3..1} ;do

#         if [ "$i" -eq 3 ] && [ "$j" -eq 3 ]; then
#             continue
#         fi
        
#         if [ "$i" -eq 3 ] && [ "$j" -eq 4 ]; then
#             continue
#         fi

xstart=$((  900000  + (i-1) * 1625000 ))
ystart=$((  5500000 - j     * 1150000 ))
xend=$((    900000  + i     * 1625000 ))
yend=$((    5500000 - (j-1) * 1150000 ))


echo $xstart

done
done
```

```python
%%bash

# Remove old merged file if it exists
rm merged_polygons.geojson 2> /dev/null

for i in {1..4..1}; do
    for j in {1..4..1}; do
        xstart=$(( 900000 + (i-1) * 1625000 ))
        ystart=$(( 5500000 - j * 1150000 ))
        xend=$(( 900000 + i * 1625000 ))
        yend=$(( 5500000 - (j-1) * 1150000 ))

        # Create a WKT (Well-Known Text) representation of the polygon
        wkt="POLYGON((${xstart} ${ystart},${xend} ${ystart},${xend} ${yend},${xstart} ${yend},${xstart} ${ystart}))"
        echo $wkt
        # Use ogr2ogr to create a GeoJSON from the WKT
        ogr2ogr -f "GeoJSON" "temp_${i}_${j}.geojson" -wrapdateline -dialect SQLITE -sql "SELECT MakePolygon(GeomFromText('${wkt}')) AS geometry"

    done
done

```

```python

# Merge individual GeoJSON files into one
ogr2ogr -f "GeoJSON" merged_polygons.geojson temp_1_1.geojson  # Use the first file to initiate the merged file

for i in {1..4..1}; do
    for j in {1..4..1}; do
        if [[ ! ($i -eq 1 && $j -eq 1) ]]; then
            ogr2ogr -f "GeoJSON" -update -append merged_polygons.geojson temp_${i}_${j}.geojson -nln merged_polygons
        fi

        # Optionally, remove individual temp files
        rm "temp_${i}_${j}.geojson"
    done
done
```
