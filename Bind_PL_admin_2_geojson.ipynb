```python
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
nuts_path = "in_data/admin/Forest_area_PL.geojson"
```

```python
admin_stats = "in_data/admin/PL_landcover.csv"
```

```python
aggregates = pd.read_csv(admin_stats)

aggregates.columns = ['NUTSID', 'cl_1_total', 'cl_2_total', 'cl_4_total']


aggregates.head()
```

```python
aggregates.columns
```

```python
features = gpd.read_file(nuts_path)
```

```python
# Set 'NUTSID' as the index in a copy of the aggregates DataFrame
aggregates_indexed = aggregates.set_index('NUTSID')

for i, col in enumerate(aggregates_indexed.columns):
    # Create a new column name for features GeoDataFrame
    label = col
    
    # Map the values from aggregates to features based on 'NUTS_ID'
    features[label] = features['NUTS_ID'].map(aggregates_indexed[col])

```

```python
features.to_file('in_data/admin/PL_stats.geojson', driver='GeoJSON')
```
