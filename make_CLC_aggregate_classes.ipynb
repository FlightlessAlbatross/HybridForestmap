```python
%%bash
output_file='/tmp/recode4bash_CLC100.txt '

for i in {1..47}; do
    # By default, set the second number to 4
    second_num=4

    # Check for special cases and change the second number accordingly
    if [ "$i" -eq 23 ] || [ "$i" -eq 24 ] || [ "$i" -eq 25 ]; then
        second_num=2
    elif [ "$i" -eq 12 ] || [ "$i" -eq 13 ] || [ "$i" -eq 14 ] || [ "$i" -eq 20 ]; then
        second_num=1
    elif [ "$i" -eq 26 ]; then
        second_num=0
    fi

    # Write to the file
    echo "$i $second_num" >> $output_file
done

echo "-128 48" >> /tmp/recode4bash_CLC100.txt 
echo "128 48" >> /tmp/recode4bash_CLC100.txt 


cat /tmp/recode4bash_CLC100.txt 
```
